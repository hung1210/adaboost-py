```python
import pandas as pandas
import numpy as numpy

data = pandas.read_csv('../input/letters-cg/letters_CG.csv')
data.head()
```

```python
data['Class'] = data['Class'].apply(lambda x: 1 if x=='C' else -1)
X = data[['x-box', 'y-box', 'width', 'high', 'onpix', 'x-bar', 'y-bar', 'x2bar', 'y2bar', 'xybar', 'x2ybr', 'xy2br', 'x-ege', 'xegvy', 'y-ege', 'yegvx']]
y = data['Class']
print(X)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.668)
print(len(X_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
clf = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
```

```python
model = clf.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
print(y_test)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
